## Encoding Ordinal variables

```python
from sklearn.preprocessing import OrdinalEncoder
cols=["Education","EnvironmentSatisfaction","JobInvolvement","JobSatisfaction","OverTime","PerformanceRating","StockOptionLevel","CommunicationSkill"]
oe=OrdinalEncoder()
data[cols]=oe.fit_transform(data[cols])
```

## Encoding Categorical variables

```python
from sklearn.preprocessing import LabelEncoder
cols=['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus']
le=LabelEncoder()
for col in cols:
    data[col]=le.fit_transform(data[col])
```

```python
data
```

## Transforming continous variables

```python
data["MonthlyIncome"]=data["MonthlyIncome"]/1000
data["EmployeeNumber"]=data["EmployeeNumber"]/1000
```

```python
cols=['Age','DistanceFromHome','MonthlyIncome','NumCompaniesWorked','PercentSalaryHike','TotalWorkingYears','TrainingTimesLastYear','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager','EmployeeNumber']
```

```python
for col in cols:
    data[col]=data[col]-data[col].mean()/data[col].std()
```

```python
data[cols].describe()
```

```python
data_train=data.iloc[0:1628,1:]
data_test=data.iloc[1628:,1:]
```

```python
data_train
```

## Hyper-parameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

#clf=GridSearchCV(svm.SVC(gamma='auto'),{
clf=GridSearchCV(LogisticRegression(),{
    'C':[7,8,9,10,12],
    #'degree':[2,3,4,5]
    #'kernel':['rbf','linear']
    }, cv=5,  return_train_score=False)
clf.fit(data_train,target)
```

```python
clf.best_params_
```

```python
model_params = {
    'svm':{
        'model': svm.SVC(gamma='auto',kernel='sigmoid'),
        'params':{
            'C':[5,10,15,20],
            'degree':[2,3,4,5]
            #'C':[3,4,5,6,7]
            #'kernel':['rbf','linear','poly', 'sigmoid']
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params':{
            'n_estimators':[50,100,200],
            'max_depth':[5,10,15],
            'min_samples_leaf':[5,6,7,8,],
            'criterion':['gini','entropy'],
            'max_features':['sqrt','log2']
        }
    },
    'logistic_regression':{
        'model': LogisticRegression(solver='saga',multi_class='auto'),
        'params':{
            'C':[3,4,5,6,7,8],
            'solver' : ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
        }
        
    }
}
```

```python
scores=[]

for model_name, mp in model_params.items():
    clf= GridSearchCV(mp['model'],mp['params'], cv=5,return_train_score=False)
    clf.fit(data_train,target)
    scores.append({
        'model': model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
```

```python
scores
```

## Using RandomForestClassifier model

```python
from sklearn.metrics import make_scorer
def ras(y,y_pred):
    return roc_auc_score(y,y_pred)
RAS=make_scorer(ras, greater_is_better=True)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score
scores = cross_val_score(RFC, data_train, target, cv=5,scoring=RAS)
scores
```
